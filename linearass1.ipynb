```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('percapita_income')
plt.scatter(df.year,df.percapita_income,color='red',marker='+')
```

```python
new_df=df.drop('percapita_income',axis='columns')
```

```python
new_df
```

```python
percapita_income=df.percapita_income
percapita_income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,percapita_income)
```

```python
reg.predict([[2020]])
```
